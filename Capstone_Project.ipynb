```python
## imported all required libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
## Read a CSV file and store it in df.

df=pd.read_csv("Nutrical Dataset.csv")

## dispalys data set with 5 rows.

df.head()
```

```python
df.describe(include='all')
```

```python
###Identify Missing Values:


missing_values = df.isnull().sum()
print(missing_values)
```

```python
df.drop_duplicates()
```

```python
#Handle missing Values

df.dropna(inplace=True)
```

```python
df
```

```python
##Analyze the distribution of calorie counts across menu items: This is done by creating a histogram chart.


plt.figure(figsize=(10, 6))
sns.histplot(df['Calories'], bins=20, kde=True, color='skyblue')
plt.title('Distribution of Calorie Counts Across Menu Items')
plt.xlabel('Calories')
plt.ylabel('Frequency')

mean_calories = np.mean(df['Calories'])
median_calories = np.median(df['Calories'])
std_dev_calories = np.std(df['Calories'])


plt.text(150, 80, f'Mean Calories: {mean_calories:.2f}', fontsize=12, color='red')
plt.text(150, 75, f'Median Calories: {median_calories:.2f}', fontsize=12, color='blue')
plt.text(150, 70, f'Standard Deviation: {std_dev_calories:.2f}', fontsize=12, color='green')

plt.show()
```

```python
## summary of Calories column

calorie_stats = df['Calories'].describe()
calorie_stats
```

```python
nutritional_summary = df[['Sugars', 'Protein', 'Carbohydrates']].describe()
nutritional_summary
```

```python
## created a boxplot to visualize the distribution of nutritional content
plt.figure(figsize=(12, 6))
sns.boxplot(x='variable', y='value', data=pd.melt(df[['Sugars', 'Protein', 'Carbohydrates']]))
plt.title('Distribution of Nutritional Content Across Items')
plt.xlabel('Nutritional Content')
plt.ylabel('Amount (grams)')
plt.show()
```

```python
###create Bar chart.


plt.figure(figsize=(12, 6))
sns.barplot(x='Category', y='Calories', data=df, ci=None, estimator='mean', palette='viridis')
plt.title('Average Calories per Food Category')
plt.xlabel('Category')
plt.ylabel('Calories')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
##According to above Bar chart, analysing the calories for provided categories, chicken and Fish have a highest calorie count in the preovided items.
```

```python
## Histogram

plt.figure(figsize=(12, 6))
sns.histplot(df['Calories'], bins=20, kde=True, color='skyblue')
plt.title('Distribution of Calorie Counts Across Menu Items')
plt.xlabel('Calories')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(10, 12))
sns.boxplot(x='Category', y='Calories', data=df, palette='Set2')
plt.title('Box Plot')
plt.xlabel('Menu Categories')
plt.ylabel('Calorie Counts')
plt.xticks(rotation=40, ha='right')
plt.show()
```
